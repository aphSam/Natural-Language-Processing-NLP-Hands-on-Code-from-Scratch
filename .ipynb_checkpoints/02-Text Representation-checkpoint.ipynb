```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

### Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["campusx watch and write comment of campusx"]).toarray()
```

```python
df1 = pd.read_csv('dataset/IMDBDataset.csv')
df1
```

```python
df1 = df1.drop(df1.index[5:])
```

```python
df1
```

```python
cv1 = CountVectorizer()
bow1 = cv1.fit_transform(df1['review'])
bow1 = cv1.transform(df1['review'])
```

```python
print(bow1[0].toarray())
print(bow1[1].toarray())
print(bow1[2].toarray())
print(bow1[3].toarray())
print(bow1[4].toarray())
```

```python
print(cv1.vocabulary_)
```

### N-gram

```python
df
```

```python
#bi-gram
cv2 = CountVectorizer(ngram_range=(2,2))
```

```python
bow2 = cv2.fit_transform(df['text'])
```

```python
print(cv2.vocabulary_)
```

```python
print(len(cv2.vocabulary_))
```

```python
print(bow2[0].toarray())
print(bow2[1].toarray())
print(bow2[2].toarray())
print(bow2[3].toarray())
```

```python
#tri-gram
cv3 = CountVectorizer(ngram_range=(3,3))
bow3 = cv3.fit_transform(df['text'])
```

```python
print(cv3.vocabulary_)
```

```python
print(len(cv3.vocabulary_))
```

```python
print(bow3[0].toarray())
print(bow3[1].toarray())
print(bow3[2].toarray())
print(bow3[3].toarray())
```

```python
#Apply on realword dataset
cv5 = CountVectorizer(ngram_range=(3,3))
```

```python
bow5 = cv5.fit_transform(df1['review'])
```

```python
print(cv5.vocabulary_)
```

```python
print(len(cv3.vocabulary_))
```

```python
print(bow5[0].toarray())
print(bow5[1].toarray())
```

```python
cv4 = CountVectorizer(ngram_range=(1,3))
bow4 = cv4.fit_transform(df['text'])
print(cv4.vocabulary_)
```

```python
print(len(cv4.vocabulary_))
```

### Tf-idf

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
print(tfidf.get_feature_names_out())
```

```python
#Apply on realword dataset
tfidf.fit_transform(df1['review']).toarray()
```

```python
print(tfidf.idf_)
```

```python
print(tfidf.get_feature_names_out())
```
